```python
from tavily import TavilyClient
from transformers import pipeline
import requests
from bs4 import BeautifulSoup
import re

# کلید API برای Tavily
tavily = TavilyClient(api_key="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")

# جستجو با استفاده از Tavily
response = tavily.search(
    query="I want to apply for Master of Autonomous systems and robots in Technical University of Darmstadt for international or non-European applicants, What is the deadline exactly?, What are the language requirements? What is exactly for language score?, What is the GPA for Bachelor's degree?, please search only from the mentioned university",
    search_depth="advanced",
    include_raw_content=True
)

# استخراج محتوا و لینک‌ها از نتایج جستجو
context = [{"url": obj["url"], "content": obj["content"]} for obj in response["results"]]

# پردازش نتایج جستجو با استفاده از Transformers
nlp = pipeline("question-answering")

# سوالات مورد نظر برای هر نتیجه جستجو
questions = [
    "What is the application deadline for international applicants?",
    "What is the language requirement for the program?",
    "What is the required language score?",
    "What is the required GPA for Bachelor's degree?"
]

answers = []

# استخراج پاسخ‌ها برای هر سوال
for qa in questions:
    for doc in context:
        url = doc['url']
        response = requests.get(url)

        if response.status_code == 200:
            webpage_content = response.text
            soup = BeautifulSoup(webpage_content, 'html.parser')

            # استخراج متن از تگ‌های مشخص شده (مثلا h1, h2, h3, p)
            extracted_text = []
            for tag in soup.find_all(['h1', 'h2', 'h3', 'p']):
                text = tag.get_text(" ", strip=True)
                extracted_text.append(text)

            relevant_text = "\n".join(extracted_text)

            # پاسخ به سوال با استفاده از Transformers
            answer = nlp(question=qa, context=relevant_text)

            # چاپ پاسخ با متن مرجع (URL)
            print(f"Question: {qa}")
            print(f"Answer with reference: {answer['answer']}")
            print(f"URL: {url}\n")

            # ذخیره پاسخ در لیست جواب‌ها
            answers.append({"question": qa, "answer": answer['answer'], "url": url})
        else:
            print(f"Failed to retrieve webpage for URL: {url}")

# چاپ پاسخ‌ها
for ans in answers:
    print(f"Question: {ans['question']}")
    print(f"Answer: {ans['answer']}")
    print(f"URL: {ans['url']}\n")
```

```python
from tavily import TavilyClient
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
from transformers import pipeline

# کلید API برای Tavily
tavily = TavilyClient(api_key="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")

# جستجو با استفاده از Tavily
query = """
    I want to apply for Master of Autonomous systems and robots in Technical University of Darmstadt for international or non-European applicants.
    What is the deadline exactly?
    What are the language requirements?
    What is exactly for language score?
    What is the GPA for Bachelor's degree?
    Please search only from the mentioned university
"""

response = tavily.search(
    query=query,
    search_depth="advanced",
    include_raw_content=True
)

# استخراج محتوا و لینک‌ها از نتایج جستجو
context = [{"url": obj["url"], "content": obj["content"]} for obj in response["results"]]

def scrape_web_content(url):
    """Scrape content from the given URL"""
    if not url:
        return "Content could not be found."

    # fetch data
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(url, headers=headers)
    if response.status_code != 200:
        return "Failed to retrieve the webpage."

    # parse result
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup

def extract_text_from_soup(soup):
    """Extract text from BeautifulSoup object"""
    page_content = []
    for tag in soup.find_all(['h1', 'h2', 'h3', 'p']):
        text = tag.get_text(" ", strip=True)
        page_content.append(text)

    # ترکیب همه عناصر به یک رشته
    page_content = "\n".join(page_content)

    # حذف همه فاصله‌ها از متن ترکیبی
    page_content = re.sub(r'\s+', ' ', page_content)

    return page_content

# تبدیل لینک‌ها به متن
text_data = ""
for entry in context:
    url = entry["url"]
    soup = scrape_web_content(url)
    text_data += extract_text_from_soup(soup)

# تقسیم داده‌ها به بخش‌های کوچکتر
chunks = [text_data[i:i+1000] for i in range(0, len(text_data), 1000)]

# استفاده از مدل زبان طبیعی متن باز
generator = pipeline('text-generation', model='EleutherAI/gpt-neo-2.7B')

# پردازش هر بخش به صورت جداگانه
responses = []
for chunk in chunks:
    llm_prompt = f"""
    Using the following information extracted from the web:
    {chunk}

    Please provide the following details in a tabular format:
    1. University Name
    2. Program Name
    3. Application Deadline
    4. Language Requirements
    5. Required Language Scores
    6. GPA Requirement for Bachelor's Degree
    """
    response = generator(llm_prompt, max_new_tokens=500, do_sample=True, temperature=0.7)
    responses.append(response[0]['generated_text'])

# ترکیب پاسخ‌ها
table_data = "\n".join(responses)

# نمایش نتیجه در قالب DataFrame
print(table_data)
```
